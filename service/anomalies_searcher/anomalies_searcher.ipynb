# Подключим модель для семантического анализа отзывов.
Информация о модели: https://huggingface.co/blanchefort/rubert-base-cased-sentiment?text=%D0%A2%D1%8B+%D0%BC%D0%BD%D0%B5+%D0%BD%D1%80%D0%B0%D0%B2%D0%B8%D1%88%D1%8C%D1%81%D1%8F.+%D0%AF+%D1%82%D0%B5%D0%B1%D1%8F+%D0%BB%D1%8E%D0%B1%D0%BB%D1%8E

```python
tokenizer = BertTokenizerFast.from_pretrained('blanchefort/rubert-base-cased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('blanchefort/rubert-base-cased-sentiment', return_dict=True)


def predict(text):
    inputs = tokenizer(text, max_length=512, padding=True, truncation=True, return_tensors='pt')
    outputs = model(**inputs)
    predicted = torch.nn.functional.softmax(outputs.logits, dim=1)
    predicted = torch.argmax(predicted, dim=1).numpy()
    return predicted
```

# Обработка заявок

```python
delta = end - start
```

```python
# Будем итерироваться по заявкам из загруженного датасета
for i in tqdm(range(delta)):

    request = data.iloc[i, :]
    
    # Критерии аномальности, которым удовлетворяет заявка
    сategories_of_anomaly = set()

    # "ИД ситуации" данной заявки. По умолчанию соответсвует её "Корневой ИД заявки"
    situation_id = request["Корневой ИД заявки"]

    # Наличие типа дефекта в Приложении 2
    request_in_appendix_2 = request['Идентификатор дефекта'] in appendix2

    # Значение поля "Повторное срок, часов" из Приложения 3
    request_appendix_3 = 0
    filt = appendix3['Корневой идентификатор'] == request['Идентификатор дефекта']
    if len(appendix3[filt]):
        request_appendix_3 = appendix3[filt].iloc[0, 3]

    # Наличие типа дефекта в Приложении 2
    request_in_appendix_4 = True
    if request['Признак дефекта возврата на доработку'] == 'Нет':
        request_in_appendix_4 = False


    # Обработка критериев 4 и 5 из ТЗ
    if request["Кол-во возвратов на доработку"] > 0:
       if request_in_appendix_4:
            сategories_of_anomaly.add(5)
    else:
       if (((request['Дата закрытия'] - request['Дата создания заявки в формате Timezone']) / pd.Timedelta(minutes=1) < 10)
           & (request["Результативность"] == "Выполнено")
           & (not request_in_appendix_2)):
           сategories_of_anomaly.add(4)


    # Обрабатка семантики отзыва и оценки качества
    if request['Признак инцидента'] != 'Да':
        if request['Отзыв'] != ' ':
            if (request['Отзыв'] == 'Работы не выполнены') or (predict(request['Отзыв'])[0] == 2):
                сategories_of_anomaly.add(6)

        if request['Оценка качества выполнения работ'] != ' ':
            if request['Оценка качества выполнения работ'] == 'Плохо':
                сategories_of_anomaly.add(6)

    # Конфигурация адреса, по котрой будет производиться поиск "схожих" заявок
    full_address = ''

    # Если в Приложении 3 в поле "Повторное срок, часов", то нам не надо рассматривать цепочку из нескольких заявок
    if request_appendix_3 == 0 or request['Признак инцидента'] == 'Да':
        pass

    # Иначе ищем ещё заявки входящие в ситуацию
    else:

        # Поле "Повторное локация" Приложения 3. 
        full_address = appendix3[filt].iloc[0, 4]
        filt2 = None

        # Сокращаем объём данных для поиска
        data_for_search2 = data[data['Индекс'] < request['Индекс']]

        # Подбираем нужный фильтр
        if full_address in ['адрес', 'Адрес']:

           filt2 = ( (data_for_search2['УНОМ'] == request['УНОМ'])
                   & (data_for_search2['Идентификатор дефекта'] == request['Идентификатор дефекта'])
                   & (data_for_search2['Признак инцидента'] == 'Нет')
                   & ((request['Дата создания заявки в формате Timezone'] -
                       data_for_search2['Дата закрытия']) / pd.Timedelta(hours=1) < request_appendix_3)
                   & (request['Дата создания заявки в формате Timezone'] > data_for_search2['Дата закрытия']))

        elif full_address == 'адрес+кв.':

           filt2 = ( (data_for_search2['УНОМ'] == request['УНОМ'])
                   & (data_for_search2['Идентификатор дефекта'] == request['Идентификатор дефекта'])
                   & (data_for_search2['Квартира'] == request['Квартира'])
                   & (data_for_search2['Признак инцидента'] == 'Нет')
                   & ((request['Дата создания заявки в формате Timezone'] - 
                       data_for_search2['Дата закрытия']) / pd.Timedelta(hours=1) < request_appendix_3)
                   & (request['Дата создания заявки в формате Timezone'] > data_for_search2['Дата закрытия']))

        elif full_address == "адрес+подъезд":

           filt2 = ( (data_for_search2['УНОМ'] == request['УНОМ'])
                   & (data_for_search2['Идентификатор дефекта'] == request['Идентификатор дефекта'])
                   & (data_for_search2['Подъезд'] == request['Подъезд'])
                   & (data_for_search2['Признак инцидента'] == 'Нет')
                   & ((request['Дата создания заявки в формате Timezone'] -
                      data_for_search2['Дата закрытия']) / pd.Timedelta(hours=1) < request_appendix_3)
                   & (request['Дата создания заявки в формате Timezone'] > data_for_search2['Дата закрытия']))

        else:

           filt2 = ( (data_for_search2['УНОМ'] == request['УНОМ'])
                   & (data_for_search2['Идентификатор дефекта'] == request['Идентификатор дефекта'])
                   & (data_for_search2['Подъезд'] == request['Подъезд'])
                   & (data_for_search2['Квартира'] == request['Квартира'])
                   & (data_for_search2['Признак инцидента'] == 'Нет')
                   & ((request['Дата создания заявки в формате Timezone'] -
                       data_for_search2['Дата закрытия']) / pd.Timedelta(hours=1) < request_appendix_3)
                   & (request['Дата создания заявки в формате Timezone'] > data_for_search2['Дата закрытия']))      

        # Подходящие под условия заявки
        requests = data_for_search2[filt2]

        # Если заявки не нашли, то эта завка и есть начальная в этой ситуации (цепочке)
        if len(requests) == 0:
            data.loc[request['Индекс'], "Начало ситуации"] = request["Дата создания заявки в формате Timezone"]

        # Иначе пересматриваем категории аномальности последней завершившейся заявки в ситуации (цепочке)
        # Критерии 1, 2, 3 в ТЗ.
        else:

            # Последняя завершившаяся заявка в ситуации
            old_request = requests.sort_values(by='Дата закрытия', ascending=False).iloc[0, :]
            old_request_index = old_request["Индекс"]
            new_categories = ''
            if old_request["Категории аномальности"] != '0':
                new_categories = old_request["Категории аномальности"]
            
            # Пересматриваем статус предыдущей заявки
            if old_request["Результативность"] == 'Выполнено':

                if old_request["Кол-во возвратов на доработку"] > 0:
                    pass

                else:

                    if ((pd.to_datetime(old_request['Дата закрытия']) - pd.to_datetime(
                            old_request['Дата создания заявки в формате Timezone'])) / pd.Timedelta(minutes=1) < 10):
                        if request_in_appendix_2:
                            new_categories += '3'

                    else:
                        if old_request["Наименование дефекта"] not in [
                           "Ввод в эксплуатацию ИПУ воды (замена, демонтаж, пропуск межповерочного интервала)",
                           "Подача документов о поверке ИПУ воды в электронном виде"]:
                            new_categories += '2'

            else:

                if ((old_request["Наименование дефекта"] not in [
                    "Ввод в эксплуатацию ИПУ воды (замена, демонтаж, пропуск межповерочного интервала)",
                    "Подача документов о поверке ИПУ воды в электронном виде"])
                        & (old_request["Вид выполненных работ"] != ['Аварийное/плановое отключение'])):
                    new_categories += '1'

            # Изменяем категории аномальности последней завершившейся заявки в ситуации (цепочке)
            categories = sorted(list(set(new_categories)))
            data.loc[old_request_index, "Категории аномальности"] = ''.join(str(x) for x in categories)


            # Ищем первичную в ситуации заявку
            fltr = (data['Корневой ИД заявки'] == old_request["ИД ситуации"])
            first_request = data[fltr]
            data.loc[first_request["Индекс"], "Окончание ситуации"] = request["Дата закрытия"]
 
            # Прикрепляем заявку у ситуации
            situation_id = old_request["ИД ситуации"]

    # Записываем категории аномальности текущей заявки и её принадлежность к ситуации.
    tmp = sorted(list(сategories_of_anomaly))
    сategories_of_anomaly = ''.join(str(x) for x in tmp)

    data.loc[request["Индекс"], "ИД ситуации"] = situation_id

    if str(сategories_of_anomaly):
        data.loc[request["Индекс"], "Категории аномальности"] = str(сategories_of_anomaly)


    # Ищем, так называемые внутренние заявки. Те, которые открылись после открытия данной, но закрылись раньше её
    filt1 = None

    # Сокращаем объём поиска
    data_for_search1 = data[data['Индекс'] > request["Индекс"]]

    # Подбираем нужный фильтр
    if full_address in ['адрес', 'Адрес']:

        filt1 = ( (data_for_search1['УНОМ'] == request['УНОМ'])
                & (data_for_search1['Идентификатор дефекта'] == request['Идентификатор дефекта'])
                & (data_for_search1['Признак инцидента'] == 'Нет')
                & (request['Дата создания заявки в формате Timezone'] < data_for_search1['Дата создания заявки в формате Timezone'])
                & (request['Дата закрытия'] > data_for_search1['Дата закрытия']))
        
    elif full_address == 'адрес+кв.':

        filt1 = ( (data_for_search1['УНОМ'] == request['УНОМ'])
                & (data_for_search1['Идентификатор дефекта'] == request['Идентификатор дефекта'])
                & (data_for_search1['Квартира'] == request['Квартира'])
                & (data_for_search1['Признак инцидента'] == 'Нет')
                & (request['Дата создания заявки в формате Timezone'] < data_for_search1['Дата создания заявки в формате Timezone'])
                & (request['Дата закрытия'] > data_for_search1['Дата закрытия']))
        
    elif full_address == "адрес+подъезд":

        filt1 = ( (data_for_search1['УНОМ'] == request['УНОМ'])
                & (data_for_search1['Подъезд'] == request['Подъезд'])
                & (data_for_search1['Признак инцидента'] == 'Нет')
                & (request['Дата создания заявки в формате Timezone'] < data_for_search1['Дата создания заявки в формате Timezone'])
                & (request['Дата закрытия'] > data_for_search1['Дата закрытия']))
        
    else:

        filt1 = ( (data_for_search1['УНОМ'] == request['УНОМ'])
                & (data_for_search1['Подъезд'] == request['Подъезд'])
                & (data_for_search1['Квартира'] == request['Квартира'])
                & (data_for_search1['Признак инцидента'] == 'Нет')
                & (request['Дата создания заявки в формате Timezone'] < data_for_search1['Дата создания заявки в формате Timezone'])
                & (request['Дата закрытия'] > data_for_search1['Дата закрытия']))
    

    # Прикрепляем внутренние заявки к той же ситуации, что и данная.
    intermediate_requests = data_for_search1[filt1]

    for request in intermediate_requests.values:
        data.loc[request[0], 'ИД ситуации'] = situation_id
        data.loc[request[0], 'Внутренняя заявка'] = 'Да'
```

# Тут мы подбираем параметры для статистичекого метода. Следующий код нужен только для обучения, но не для разметки заявок по аномальности.

```python
identifiers = appendix3['Корневой идентификатор'].values
identifiers_parametr = list()

for identifier in tqdm(identifiers):
    filtr_id = (identifier == situations['Идентификатор дефекта'])
    data_id = situations[filtr_id]
    if len(data_id) > 50:
        IQR = data_id["Продолжительность ситуации"].quantile(q=0.75) - data_id["Продолжительность ситуации"].quantile(q=0.25)
        up_bound = data_id["Продолжительность ситуации"].quantile(q=0.75) + 8*IQR
        identifiers_parametr.append(up_bound)
    else:
        identifiers_parametr.append(0.)
```

# Поиск аномалии 7 (долгое решение ситуации)

Следующий словарь может также быть выгружен из  'Appendix3_with_percentiles.csv'

```python
identifiers_parametr_dict = dict()
for i in range(len(identifiers)):
    identifiers_parametr_dict[identifiers[i]] = identifiers_parametr[i]
```

Аномалия 7 присуждается ситуациям, которые уже имеют какую-то другую аномалию. Она поможет приоритезировать заявки, которые выполняются дольше, чем обычно.

```python
for i in tqdm(range(delta)):
    index = data.iloc[i, :]["Индекс"]
    if data.loc[index, "Корневой ИД заявки"] == data.loc[index, "ИД ситуации"]:
        if data.loc[index, 'Идентификатор дефекта'] in identifiers_parametr_dict.keys():
            up_bound = identifiers_parametr_dict[data.loc[index, 'Идентификатор дефекта']]
            if  (up_bound > 0.01) & ((data.loc[index, 'Окончание ситуации'] - data.loc[index, 'Начало ситуации']) / pd.Timedelta(hours=1) > up_bound):
                categories = set(data.loc[index, "Категории аномальности"])
                if not '0' in categories:
                    categories.add('7')
                    categories_list_int = sorted([int(x) for x in categories])
                    categories_list_str = [str(x) for x in categories_list_int]
                    data.loc[index, "Категории аномальности"] = ''.join(categories_list_str)
```
